```python
import pandas as pd
import numpy as np
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)

```

```python
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
ratings
```

```python
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
akas
```

## Required Preprocessing - Details

    Filtering/Cleaning Steps:
        Title Basics:
            Replace "\N" with np.nan
            Eliminate movies that are null for runtimeMinutes
            Eliminate movies that are null for genre
            keep only titleType==Movie
            keep startYear 2000-2022
            Eliminate movies that include "Documentary" in genre (see tip below)
            Keep only US movies (Use AKAs table, see "Filtering one dataframe based on another" section below)

    AKAs:
        keep only US movies.
        Replace "\N" with np.nan
    Ratings:
        Replace "\N" with np.nan (if any)
        Keep only US movies (Use AKAs table, see "Filtering one dataframe based on another" section below)

### akas

```python
akas = akas.replace({'\\N':np.nan})
```

```python
akas
```

```python
# keep only US movies.
akas_us = akas['region']=='US'
```

```python
akas = akas[akas_us]
akas
```

```python
akas_final = akas.reset_index(drop=True)
akas_final
```

### basics

```python
basics = basics.replace({'\\N':np.nan})
```

```python
basics
```

```python
basics = basics.drop_duplicates()
```

```python
## Eliminate movies that are null for runtimeMinutes
basics.dropna(subset=['runtimeMinutes'], inplace=True)
```

```python
basics
```

```python
# Eliminate movies that are null for genre
basics.dropna(subset=['genres'], inplace=True)
```

```python
basics['titleType'].unique()
```

```python
## keep only titleType==Movie
movie_Filter = basics['titleType']=='movie'
basics[movie_Filter]
```

```python
basics = basics[movie_Filter]
```

```python
basics.reset_index(drop=True)
```

```python
basics.info()
```

```python
## keep startYear 2000-2022
movie_startYear = basics['startYear']>'1999'
movie_startYear
```

```python
basics = basics[movie_startYear]
```

```python
basics
```

```python
## Eliminate movies that include "Documentary" in genre
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
basics
```

```python
# Keep only US movies 
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```

```python
basics_final = basics.reset_index(drop=True)
basics_final.info()
```

### Ratings

```python
ratings = ratings.replace({'\\N':np.nan})
```

```python
ratings
```

```python
keepers_ratings = ratings['tconst'].isin(akas['titleId'])
keepers_ratings
```

```python
ratings =ratings[keepers_ratings]
```

```python
ratings_final = ratings.reset_index(drop=True)
ratings_final
```

### Saving as Data

```python
# example making new folder with os
import os
os.makedirs('Data/',exist_ok=True) 
# Confirm folder created
os.listdir("Data/")
```

```python
## Save current dataframe to file.
basics_final.to_csv("Data/final_basics.csv.gz",compression='gzip',index=False)

```

```python
akas_final.to_csv("Data/final_akas.csv.gz",compression='gzip',index=False)
ratings_final.to_csv("Data/final_ratings.csv.gz",compression='gzip',index=False)
```
